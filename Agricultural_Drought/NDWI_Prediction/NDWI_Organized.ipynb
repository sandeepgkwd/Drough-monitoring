## 1. Google Earth Engine Authentication

```python
# Import libraries and authenticate
import ee
import geemap
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

# Authenticate and initialize Earth Engine
ee.Authenticate()
ee.Initialize(project='your-project-id')  # Replace with your project ID
```

## 2. Region of Interest (ROI) Definition

```python
# Define study area coordinates
coords = [
    [74.92414740325971, 19.670854004345237],
    [75.20567206146283, 19.670854004345237],
    [75.20567206146283, 19.832415852455963],
    [74.92414740325971, 19.832415852455963],
    [74.92414740325971, 19.670854004345237]
]
roi = ee.Geometry.Polygon(coords)

# Create interactive map
Map = geemap.Map(center=[19.75, 75.065], zoom=12)
Map.add_basemap("SATELLITE")
Map.addLayer(roi, {"color": "red"}, "Study Area")
Map
```

## 3. Historical Data Extraction

```python
# Load MODIS Surface Reflectance data
start_date = '2000-01-01'
end_date = '2025-12-31'

modis_collection = ee.ImageCollection('MODIS/061/MOD09A1') \
    .filterDate(start_date, end_date) \
    .filterBounds(roi) \
    .select(['sur_refl_b04', 'sur_refl_b02'])  # Green and NIR bands

# Function to calculate NDWI
def add_ndwi(image):
    ndwi = image.normalizedDifference(['sur_refl_b04', 'sur_refl_b02']).rename('NDWI')
    return image.addBands(ndwi)

# Apply NDWI calculation
modis_with_ndwi = modis_collection.map(add_ndwi)

print(f"Total images in collection: {modis_collection.size().getInfo()}")
```

```python
# Extract time series data
def get_mean_ndwi(image):
    mean_value = image.select('NDWI').reduceRegion(
        reducer=ee.Reducer.mean(),
        geometry=roi,
        scale=500
    ).get('NDWI')
    date = image.date().format('YYYY-MM-dd')
    return ee.Feature(None, {'date': date, 'mean_ndwi': mean_value})

# Extract time series
ndwi_list = modis_with_ndwi.map(get_mean_ndwi).getInfo()

# Convert to DataFrame
records = []
for f in ndwi_list['features']:
    props = f['properties']
    date = props.get('date')
    ndwi_value = props.get('mean_ndwi', None)
    records.append({'date': date, 'ndwi': ndwi_value})

df = pd.DataFrame(records)
df['date'] = pd.to_datetime(df['date'])
df['ndwi'] = pd.to_numeric(df['ndwi'], errors='coerce')
df = df.sort_values('date').reset_index(drop=True)

print(f"Extracted {len(df)} data points")
print(f"Date range: {df['date'].min()} to {df['date'].max()}")
```

## 4. Data Processing and Cleaning

```python
# Data cleaning and preprocessing
print(f"Missing values before cleaning: {df['ndwi'].isnull().sum()}")

# Interpolate missing values
df['ndwi'].interpolate(method='linear', inplace=True)
df['ndwi'].fillna(method='bfill', inplace=True)
df['ndwi'].fillna(method='ffill', inplace=True)

# Remove outliers using quantile-based capping
lower_bound = df['ndwi'].quantile(0.05)
upper_bound = df['ndwi'].quantile(0.95)
df['ndwi'] = df['ndwi'].clip(lower=lower_bound, upper=upper_bound)

# Apply smoothing
df['ndwi_smoothed'] = df['ndwi'].rolling(window=3, center=True).mean()
df['ndwi_smoothed'].fillna(method='bfill', inplace=True)
df['ndwi_smoothed'].fillna(method='ffill', inplace=True)

print(f"Missing values after cleaning: {df['ndwi'].isnull().sum()}")
print(f"NDWI range: {df['ndwi'].min():.4f} to {df['ndwi'].max():.4f}")
```

## 5. Historical Data Visualization

```python
# Create comprehensive time series plot
plt.style.use('seaborn-v0_8-whitegrid')
fig, ax = plt.subplots(figsize=(15, 8))

# Plot original and smoothed data
ax.plot(df['date'], df['ndwi'], alpha=0.3, color='blue', label='Original NDWI')
ax.plot(df['date'], df['ndwi_smoothed'], color='green', linewidth=2, label='Smoothed NDWI')

ax.set_title('Historical NDWI Time Series (2000-2025)', fontsize=16, fontweight='bold')
ax.set_xlabel('Date', fontsize=12)
ax.set_ylabel('NDWI Value', fontsize=12)
ax.legend()
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Display basic statistics
print("\nNDWI Statistics:")
print(df['ndwi_smoothed'].describe())
```

## 6. LSTM Model Development

```python
# Import deep learning libraries
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Set device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

# Prepare data for LSTM
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(df[['ndwi_smoothed']])

# Create sequences
def create_sequences(data, seq_length):
    X, y = [], []
    for i in range(seq_length, len(data)):
        X.append(data[i-seq_length:i, 0])
        y.append(data[i, 0])
    return np.array(X), np.array(y)

seq_length = 12  # Use 12 time steps to predict next value
X, y = create_sequences(scaled_data, seq_length)

# Train-test split
split_idx = int(0.8 * len(X))
X_train, X_test = X[:split_idx], X[split_idx:]
y_train, y_test = y[:split_idx], y[split_idx:]

# Convert to tensors
X_train = torch.FloatTensor(X_train).unsqueeze(-1).to(device)
y_train = torch.FloatTensor(y_train).to(device)
X_test = torch.FloatTensor(X_test).unsqueeze(-1).to(device)
y_test = torch.FloatTensor(y_test).to(device)

print(f"Training data shape: {X_train.shape}")
print(f"Test data shape: {X_test.shape}")
```

```python
# Define LSTM model
class NDWIPredictor(nn.Module):
    def __init__(self, input_size=1, hidden_size=64, num_layers=2, dropout=0.2):
        super(NDWIPredictor, self).__init__()
        self.hidden_size = hidden_size
        self.num_layers = num_layers
        
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, dropout=dropout, batch_first=True)
        self.fc = nn.Linear(hidden_size, 1)
        
    def forward(self, x):
        lstm_out, _ = self.lstm(x)
        return self.fc(lstm_out[:, -1, :])

# Initialize model
model = NDWIPredictor().to(device)
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Training loop
model.train()
for epoch in range(100):
    optimizer.zero_grad()
    outputs = model(X_train)
    loss = criterion(outputs.squeeze(), y_train)
    loss.backward()
    optimizer.step()
    
    if (epoch + 1) % 20 == 0:
        print(f'Epoch [{epoch+1}/100], Loss: {loss.item():.6f}')s, 
                           batch_first=True, dropout=dropout)
        self.dropout = nn.Dropout(dropout)
        self.fc1 = nn.Linear(hidden_size, hidden_size // 2)
        self.fc2 = nn.Linear(hidden_size // 2, 1)
        self.relu = nn.ReLU()
        
    def forward(self, x):
        lstm_out, _ = self.lstm(x)
        last_output = lstm_out[:, -1, :]
        dropped = self.dropout(last_output)
        fc1_out = self.relu(self.fc1(dropped))
        fc1_out = self.dropout(fc1_out)
        output = self.fc2(fc1_out)
        return output

# Initialize model
model = NDWIPredictor().to(device)
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

print(f"Model parameters: {sum(p.numel() for p in model.parameters()):,}")
print(model)
```

```python
# Train the model
num_epochs = 100
train_losses = []
val_losses = []

model.train()
for epoch in range(num_epochs):
    # Training
    optimizer.zero_grad()
    outputs = model(X_train)
    loss = criterion(outputs.squeeze(), y_train)
    loss.backward()
    optimizer.step()
    
    train_losses.append(loss.item())
    
    # Validation
    model.eval()
    with torch.no_grad():
        val_outputs = model(X_test)
        val_loss = criterion(val_outputs.squeeze(), y_test)
        val_losses.append(val_loss.item())
    
    model.train()
    
    if (epoch + 1) % 20 == 0:
        print(f'Epoch [{epoch+1}/{num_epochs}], Train Loss: {loss.item():.6f}, Val Loss: {val_loss.item():.6f}')

print("Training completed!")
```

## 7. Model Evaluation and Predictions

```python
# Evaluate model performance
model.eval()
with torch.no_grad():
    train_pred = model(X_train).cpu().numpy()
    test_pred = model(X_test).cpu().numpy()

# Inverse transform predictions
train_pred = scaler.inverse_transform(train_pred)
test_pred = scaler.inverse_transform(test_pred)
y_train_actual = scaler.inverse_transform(y_train.cpu().numpy().reshape(-1, 1))
y_test_actual = scaler.inverse_transform(y_test.cpu().numpy().reshape(-1, 1))

# Calculate metrics
train_mse = mean_squared_error(y_train_actual, train_pred)
test_mse = mean_squared_error(y_test_actual, test_pred)
train_mae = mean_absolute_error(y_train_actual, train_pred)
test_mae = mean_absolute_error(y_test_actual, test_pred)
train_r2 = r2_score(y_train_actual, train_pred)
test_r2 = r2_score(y_test_actual, test_pred)

print("\n=== Model Performance ===")
print(f"Training - MSE: {train_mse:.6f}, MAE: {train_mae:.6f}, R²: {train_r2:.4f}")
print(f"Testing  - MSE: {test_mse:.6f}, MAE: {test_mae:.6f}, R²: {test_r2:.4f}")
```

```python
# Generate future predictions
def predict_future(model, last_sequence, n_future, scaler):
    model.eval()
    predictions = []
    current_seq = last_sequence.clone()
    
    with torch.no_grad():
        for _ in range(n_future):
            pred = model(current_seq.unsqueeze(0).unsqueeze(-1))
            predictions.append(pred.item())
            
            # Update sequence
            current_seq = torch.cat([current_seq[1:], pred.squeeze()])
    
    # Inverse transform
    predictions = np.array(predictions).reshape(-1, 1)
    predictions = scaler.inverse_transform(predictions)
    
    return predictions.flatten()

# Predict next 24 months (2026-2027)
last_sequence = torch.FloatTensor(scaled_data[-seq_length:, 0]).to(device)
future_predictions = predict_future(model, last_sequence, 24, scaler)

# Create future dates
last_date = df['date'].iloc[-1]
future_dates = pd.date_range(start=last_date + pd.DateOffset(months=1), periods=24, freq='M')

print(f"\nGenerated {len(future_predictions)} future predictions")
print(f"Prediction range: {future_dates[0].strftime('%Y-%m')} to {future_dates[-1].strftime('%Y-%m')}")
```
